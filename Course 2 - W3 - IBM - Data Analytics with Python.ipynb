# GroupBy in Python (Agrupamento de dados)


```python
# digamos que estamos interessados em encontrar o preço médio dos veículos e observar
# como eles diferem entre diferentes tipos de estilos de carroçaria (body-style) e
# variáveis de drive-wheels (rodas motrizes).

# primeiro escolhemos as três colunas de dados que estamos interessados, o que é feito na primeira linha de código.


df_test = df[['drive-wheels', 'body-style', 'price']]
```

```python
df_test
```

```python
# Em seguida, agrupamos os dados reduzidos de acordo com as rodas motrizes (drive-wheels) e
# o estilo do corpo (body-style) na segunda linha.

df_grp = df_test.groupby(['drive-wheels', 'body-style', 'price'], as_index = False).mean()
```

```python
df_grp
```

```python
# sando o método .pivot do Panda, podemos girar a variável de estilo do corpo
# para que ela seja exibida ao longo das colunas e as rodas motrizes serão exibidas ao longo das linhas.


df_pivot = df_grp.pivot(index='drive-wheels', columns='body-style')
```
